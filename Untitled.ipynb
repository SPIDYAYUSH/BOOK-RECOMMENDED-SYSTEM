## popularity based recomdiation



```python
ratings_with_name =ratings.merge(books,on='ISBN')
```

```python
num_rating_df= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df= ratings_with_name.groupby(['Book-Title']) ['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on ='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_rating']>250].sort_values('avg_rating' , ascending=False).head(50)
```

```python
popular_df.merge(books,on='Book-Title')
```

```python
popular_df = popular_df.merge(books, on ='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

 COLLABORATIVE FILTERING BASE DRECOMMEDER SYSTEM
 

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
x
padhe_likhe_users

                                  
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating

                                  
```

```python
y =filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
famous_books = y[y].index
y
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID' , values ='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0] 
    similarity_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse = True)
    

    data = []
    for i in similarity_items:
        item = []
        temp =  books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data 
        
        
    
```

```python
recommend("1984")
```
